# Task 1: Write Sales Records to a File

```python


# Step 1: Create a list of sales
sales = [1200, 450, 980, 1500, 3000]


# Step 2: Open the file in write mode
file = open("sales_data.txt", "w")


# Step 3: Write each sale into the file
for sale in sales:
    file.write(str(sale) + "\n")


# Step 4: Close the file
file.close()


# Step 5: Reopen the file in read mode
file = open("sales_data.txt", "r")


# Step 6: Print file contents
print("Sales Records:\n")
print(file.read())


# Step 7: Close the file
file.close()
```

# Extra (Optional): Store Sales in One Line (Comma-Separated)

```python
# Sales List
sales = [1200, 450, 980, 1500, 3000]

# Open file
file = open("sales_data.txt", "w")

# Write comma-separated values
for i in range(len(sales)):

    file.write(str(sales[i]))

    if i != len(sales) - 1:
        file.write(",")

# Close file
file.close()

# Read file
file = open("sales_data.txt", "r")
print(file.read())
file.close()
```

# Task 2: Read File in Different Ways

```python


# -------------------------------
# 1. Read the entire file
# -------------------------------

file = open("sales_data.txt", "r")

print("Complete File:\n")
print(file.read())

file.close()


# -------------------------------
# 2. Read only the first line
# -------------------------------

file = open("sales_data.txt", "r")

print("First Line:")
print(file.readline())

file.close()


# -------------------------------
# 3. Read all lines into a list
# -------------------------------

file = open("sales_data.txt", "r")

lines = file.readlines()

file.close()

print("List of Lines:")
print(lines)


# -------------------------------
# 4. Convert to Integer List
# -------------------------------

sales = []

for line in lines:

    line = line.strip()      # Remove '\n'

    sales.append(int(line))

print("\nSales as Integer List:")
print(sales)
```

# Task 3: Append New Sales.

```python


# New sales to append
new_sales = [5000, 2500, 1700]

# Step 1: Open file in append mode
file = open("sales_data.txt", "a")

# Step 2: Append each sale on a new line
for sale in new_sales:
    file.write(str(sale) + "\n")

# Step 3: Close the file
file.close()


# ----------------------------------------------------------
# Read the updated file
# ----------------------------------------------------------

file = open("sales_data.txt", "r")

print("Updated Sales Records:\n")
print(file.read())

file.close()


# ----------------------------------------------------------
# Extra: Count total number of lines
# ----------------------------------------------------------

file = open("sales_data.txt", "r")

lines = file.readlines()

print("Total Number of Sales =", len(lines))

file.close()
```

# Task 4: Generate Summary Report.

```python


# Step 1 : Open the sales file
file = open("sales_data.txt", "r")

# Step 2 : Read all lines
lines = file.readlines()

file.close()

# Step 3 : Convert lines into integer list
sales = []

for line in lines:
    sales.append(int(line.strip()))

# Step 4 : Calculate summary
total_sales = sum(sales)
average_sales = total_sales / len(sales)
highest_sale = max(sales)
lowest_sale = min(sales)

# Step 5 : Open summary file in write mode
summary = open("sales_summary.txt", "w")

summary.write("===== SALES SUMMARY =====\n\n")
summary.write(f"Total Sales : {total_sales}\n")
summary.write(f"Average Sales : {average_sales:.2f}\n")
summary.write(f"Highest Sale : {highest_sale}\n")
summary.write(f"Lowest Sale : {lowest_sale}\n")

summary.close()

# Step 6 : Read and display summary file
summary = open("sales_summary.txt", "r")

print(summary.read())

summary.close()
```

# Task 5: Product Info File (User Input)

# Task 6: Read File Safely (Exception Handling).

```python

# Step 1 : Ask user for file name
file_name = input("Enter File Name : ")

try:

    # Step 2 : Open the file
    file = open(file_name, "r")

    # Step 3 : Read the contents
    print("\nFile Contents:\n")
    print(file.read())

    # Step 4 : Close the file
    file.close()

# Step 5 : Handle file not found error
except FileNotFoundError:

    print("Error: File does not exist.")

# Step 6 : Handle any other unexpected errors
except Exception as e:

    print("Unexpected Error :", e)
```

# Task 7: Mini Project – Export Discounted Prices.

```python


# Step 1 : Open the sales data file
file = open("sales_data.txt", "r")

# Step 2 : Read all lines
lines = file.readlines()

# Step 3 : Close the file
file.close()

# Step 4 : Convert strings into integers
sales = []

for line in lines:
    sales.append(int(line.strip()))

# Step 5 : Apply 10% discount
discounted_sales = []

for price in sales:

    discount_price = price - (price * 10 / 100)

    discounted_sales.append(round(discount_price, 2))

# Step 6 : Open a new file in write mode
file = open("discounted_sales.txt", "w")

# Step 7 : Write discounted prices
for price in discounted_sales:

    file.write(str(price) + "\n")

# Step 8 : Close the file
file.close()

# Step 9 : Read and display the new file
file = open("discounted_sales.txt", "r")

print("Discounted Sales:\n")
print(file.read())

# Step 10 : Close the file
file.close()
```
