# Task 1: Discount Rules (if / elif / else)

```python
# Task 1: Discount Calculator

try:
    # Taking input from user
    order_amount = float(input("Enter Order Amount: "))

    # Applying discount rules
    if order_amount >= 2000:
        discount = order_amount * 0.15

    elif order_amount >= 1500:
        discount = order_amount * 0.10

    elif order_amount >= 1000:
        discount = order_amount * 0.07

    else:
        discount = 0

    # Price after discount
    subtotal = order_amount - discount

    # Extra: 5% Tax
    tax = subtotal * 0.05

    # Final amount
    final_amount = subtotal + tax

    # Output
    print("\n----- Bill -----")
    print("Order Amount :", order_amount)
    print("Discount     :", discount)
    print("Subtotal     :", subtotal)
    print("Tax (5%)     :", tax)
    print("Final Amount :", final_amount)

except ValueError:
    print("Invalid Input! Please enter a numeric value.")
```

# Task 2: Process Multiple Orders (for loop)

```python
# Task 2: Process Multiple Orders

orders = [1200, 2500, 800, 1750, 3000]

total_revenue = 0
discount_orders = 0

print("Order\tDiscount\tFinal Amount")
print("-" * 35)

for amount in orders:

    # Discount calculation
    if amount >= 2000:
        discount = amount * 0.15

    elif amount >= 1500:
        discount = amount * 0.10

    elif amount >= 1000:
        discount = amount * 0.07

    else:
        discount = 0

    final_amount = amount - discount

    print(amount, "\t", discount, "\t\t", final_amount)

    total_revenue += final_amount

    if discount > 0:
        discount_orders += 1

print("\nTotal Revenue :", total_revenue)
print("Orders with Discount :", discount_orders)
```

# Task 3: User Menu (while loop + break/continue)

```python
# Task 3: User Menu

orders = []

while True:

    print("\n===== MENU =====")
    print("1 - Add Order Amount")
    print("2 - Show Orders")
    print("q - Quit")

    choice = input("Enter Choice: ")

    if choice == "1":

        try:
            amount = float(input("Enter Order Amount: "))
            orders.append(amount)
            print("Order Added Successfully!")

        except ValueError:
            print("Invalid Amount!")
            continue

    elif choice == "2":

        if len(orders) == 0:
            print("No Orders Available.")
            continue

        total = 0

        print("\nOrder\tDiscount\tFinal")

        for amount in orders:

            if amount >= 2000:
                discount = amount * 0.15

            elif amount >= 1500:
                discount = amount * 0.10

            elif amount >= 1000:
                discount = amount * 0.07

            else:
                discount = 0

            final = amount - discount
            total += final

            print(amount, "\t", discount, "\t\t", final)

        print("Total Revenue:", total)

    elif choice.lower() == "q":
        print("Program Closed.")
        break

    else:
        print("Invalid Choice!")
        continue
```

# Task 4: Loop Control (break & continue)


```python
# Task 5: Loop Control

daily_sales = [200, 150, 0, 400, 50, -1, 300]

total_sales = 0

for sale in daily_sales:

    # Stop processing if corrupted data
    if sale == -1:
        print("Corrupted Data Found!")
        break

    # Skip zero sales
    if sale == 0:
        continue

    total_sales += sale

print("Total Sales :", total_sales)
```
