# TASK - 3 

//# Dictionary with product prices


```python
price_dict = {
    "Laptop": 60000,
    "Mouse": 500,
    "Keyboard": 1200,
    "Monitor": 10000,
    "Printer": 9000,
    "Headphones": 2500
}

print("Original Dictionary:")
print(price_dict)
```

//# Add new product


```python
price_dict["Webcam"] = 3000
```

// # Update price


```python
price_dict["Mouse"] = 650
```

// # Remove a product safely


```python
remove_product = "Printer"

removed = price_dict.pop(remove_product, None)

if removed is not None:
    print(f"\n'{remove_product}' removed successfully.")
else:
    print(f"\n'{remove_product}' not found.")
```

// # Print average price


```python
average_price = sum(price_dict.values()) / len(price_dict)

print("\nAverage Price =", average_price)
```

//# Extra: Maximum and Minimum price


```python
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)

print("\nHighest Price Product:")
print(max_product, "=", price_dict[max_product])

print("\nLowest Price Product:")
print(min_product, "=", price_dict[min_product])
```

# TASK - 4

// # Catalog using tuples


```python
catalog = [
    ("Laptop", 60000, "Electronics"),
    ("Mouse", 650, "Accessories"),
    ("Keyboard", 1200, "Accessories"),
    ("Monitor", 10000, "Electronics"),
    ("Headphones", 2500, "Audio"),
    ("Webcam", 3000, "Accessories")
]

print("Catalog:")
for item in catalog:
    print(item)
```

// # Create category to products dictionary


```python
category_to_products = {}

for product, price, category in catalog:

    if category not in category_to_products:
        category_to_products[category] = []

    category_to_products[category].append(product)

print("\nCategory to Products Dictionary:")

for category, items in category_to_products.items():
    print(category, ":", items)
```

// # Find category having maximum products


```python
max_category = max(category_to_products, key=lambda x: len(category_to_products[x]))

print("\nCategory with Maximum Products:", max_category)

print("Products:")

for product in category_to_products[max_category]:
    print(product)
```
